Это координаты центров регионов Игры с разбивкой по областям (Area) - для последующей генерации Beast и hostile


```python
df=pd.read_csv('/content/testfile.txt', names = ['Цвет','ячейка', 'координаты'])
```

```python
df
```

```python
df.loc[df['Цвет']==1, 'Зона']='Вода'
df.loc[df['Цвет']==2, 'Зона']='никакая'
df.loc[df['Цвет']==3, 'Зона']='Побережье'
df.loc[df['Цвет']==4, 'Зона']='Горы'
df.loc[df['Цвет']==5, 'Зона']='Пустыня'
df.loc[df['Цвет']==6, 'Зона']='Леса'
df.loc[df['Цвет']==7, 'Зона']='Поле'
```

```python
df['x'] = df['координаты'].apply(lambda word: word.split(';')[0])
df['y'] = df['координаты'].apply(lambda word: word.split(';')[1])
```

```python
df['Строка'] = df['ячейка'].apply(lambda word: word[1:2])
df['Столбец'] = df['ячейка'].apply(lambda word: word[3:4])
```
